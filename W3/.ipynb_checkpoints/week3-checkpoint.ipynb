```python
id_name_dict = {
    128: 'Mary-Kate', 39: 'Ashley', 207: 'Lindsey', 127: 'Naomi', 237: 'Avril',
    33: 'Bob', 50: 'Bill', 31: 'John', 100: 'Dick', 241: 'Ed'
}
df10['user_id'] = df10['user_id'].map(id_name_dict)
```

```python
color_dic = {
    'Mary-Kate': 'pink', 'Ashley': 'darkviolet', 'Lindsey':'blueviolet', 
    'Naomi': 'hotpink', 'Avril': 'orchid', 
    'Bob': 'firebrick', 'Bill': 'gold', 'John': 'forestgreen', 
    'Dick': 'slategrey', 'Ed':'brown'
}
```

Гистограмма распределения длины сессии в секундах (session_timespan):

```python
plt.figure(figsize=(16, 6))
plt.xlim(0, 200)
sns.distplot(df10[df10['session_timespan'] <= 200]['session_timespan'], bins=50, color='darkviolet')
plt.xlabel('длина сессии в секундах')
plt.ylabel('доля сессий с данной длиной')
```

Гистограммы распределения длины сессии в секундах (session_timespan) для каждого из 10 пользователей по отдельности:

```python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(16, 10))

for i, (user, user_df) in enumerate(df10.groupby('user_id')):
    ax = axes[i % 5, i // 5]
    ax.set_title(user)
    ax.set_xlim(0, 200)
    sns.distplot(
        user_df[user_df['session_timespan'] <= 200]['session_timespan'], bins=50,
        color=color_dic[user], ax=ax
    )
    ax.set_xlabel('длина сессии в секундах')
    ax.set_ylabel('доля сессий')

fig.tight_layout()
```

# Часть 3

Добавим новые признаки для каждой сессии:

* Количество строк 'youtube' со всех сайтах сессии (#youtube);
* Количество строк 'google' во всех сайтах сессии (#google);
* Количестов строк 'facebook' во всех сайтах сессии (#facebook);
* Девять признаков time_diff1, time_diff2..., time_diff9 -- количество секунд между переходом от первого сайта ко второму, от второго к третьему, ..., от предпоследнего к последнему;
* Начало сессии (day_part): утром (до 10:00, метка 1), днём (с 10:00 до 15:00, метка 2), во второй половине дня (с 15:00 до 19:00, метка 3), вечером (с 19:00 до 00:00, метка 4). Можно было бы сделать четыре бинарных признака, если бы этот признак был бы по-настоящему категориальным. Однако разные его значения имеет смысл сравнивать между собой (для двух сессий 3 < 4 означает, что вторая сессия началась позже первой). Интуиция в том, что час начала сессии не очень информативен. Наверное, более информативным должно быть то, что одни пользователи предпочитают посидеть в Интрнете утром, а другие -- вечером;
* Средняя длина названия сайта в рамках одной сессии (avg_site_len). Интуиция: одни пользователи посещают обычные сайты, длину которых обчыно делают короткой и запоминающейся. Другие пользователи посещают необычные сайты, длина названия которых может быть большой (например, топ-7 и топ-8 в списке самых популярных сайтов).

Какие признаки не будем добавлять и почему:

* На первый взгляд последняя часть доменного имени должна быть информативной (наверное, пользователи из России чаще посещают сайты, оканчивающиеся на .ru, чем пользователи из Франции). Однако, судя по всему, данные собраны по французским пользователям, потому что домен .fr явно преобладает над другими национальными доменами.

* Год. Данных слишком мало. Если бы данных было много (например, за 20+ лет подряд), то, наверное, можно было бы предполагать, что с годами интересы пользователей меняются. Но они вряд ли меняются на протяжении нескольких месяцев.

* Месяц. Аналогично ситуации с годом -- данных недостаточно.

* День месяца. Кажется, что день недели должен быть гораздо информативнее.

```python
for folder in ['3users', '10users', '150users']:
    fe_df = utils.prepare_train_set_with_fe(
        path_to_csv_files=os.path.join(PATH_TO_DATA, folder),
        site_freq_path=os.path.join('..', 'week1', 'site_freq_' + folder + '.pickle'),
        session_length=10,
        window_size=10
    )
    with open('feature_engineering_' + folder + '.pickle', 'wb') as fe_file:
        pickle.dump(fe_df, fe_file)
```
